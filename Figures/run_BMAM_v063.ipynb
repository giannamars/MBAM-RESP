## MBAM Iterations
This script runs the zeroth model MBAM iteration and the subsequent iterations sepataretly (~1 hour)

```julia
# Load the heterotrophic respiration functions, the x_grid(Temp, Time) that we evaluate over and the initial condition phi0
include("../BK_functions/bk_setup_script.jl"); 

using JLD
```

```julia
# Run MBAM on original model
outp = MBAM(BK_simulator, phi0, x_grid, model_iters=[51],
            move_dir=[-1], boundary_time=3, verbose=1, maxEV = 1e-20, maxVnorm = [5e-2, 1e-5], reduce_func=BK_reduce); 
save("results/MBAM_outp_orig.jld", "outp", outp)
```

```julia
# Load the once reduced functions, the x_grid(Temp, Time) that we evaluate over and the initial condition phi1
include("../BK_functions/bk_setup_script1.jl"); 
```

```julia
outp = MBAM(BK_simulator, phi0, x_grid, model_iters=[10,11,12,13],
            move_dir=[1,1,-1,1], boundary_time=3, verbose=1, maxEV = 1e-12, maxVnorm = [5e-2, 1e-5], reduce_func=BK_reduce); 
save("results/MBAM_outp_final.jld", "outp", outp)
```
